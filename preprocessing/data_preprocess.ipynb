```python
df = []

for i in range(len(captures)):
    df.append(pd.read_table(filepath_or_buffer = captures[i], skiprows=10, nrows= 100000))
    df[i].columns=['ts',
        'uid',
        'id.orig_h',
        'id.orig_p',
        'id.resp_h',
        'id.resp_p',
        'proto',
        'service',
        'duration',
        'orig_bytes',
        'resp_bytes',
        'conn_state',
        'local_orig',
        'local_resp',
        'missed_bytes',
        'history',
        'orig_pkts',
        'orig_ip_bytes',
        'resp_pkts',
        'resp_ip_bytes',
        'label']
    df[i].drop(df[i].tail(1).index, inplace=True)
```

```python
frames = []
for i in range(len(df)):
    frames.append(df[i])
df_c = pd.concat(frames)
df_c
```
